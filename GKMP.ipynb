```python
import matplotlib.pyplot as plt
import myokit

# Get model and protocol, create simulation
m,p,x = myokit.load('untitled.mmt')
def make_protocol():
    p = myokit.Protocol()
    p.schedule(1, 500, 5, period=0, multiplier=0)
    
    return p  
p = make_protocol()
s = myokit.Simulation(m, p)

# Run simulation
d = s.run(1000)

# Display the results
var = 'Cai.Cai'
plt.figure()
plt.title('Membrane potential')
plt.plot(d['environment.time'], d[var])
plt.ylabel('Ca_i')
plt.xlabel('Time ( ms)')
plt.show()
```

```python
m,p,x = myokit.load('new-model.mmt')
def make_protocol():
    p = myokit.Protocol()
    p.schedule(1, 0, 2, period=0, multiplier=0)
    
    return p  
p = make_protocol()
s = myokit.Simulation(m, p)

# Run simulation
d = s.run(1000)

# Display the results
var = 'membrane.i_Stim'
plt.figure()
plt.title('Membrane potential')
plt.plot(d['environment.time'][140:150], d[var][140:150])
plt.ylabel('Ca_i')
plt.xlabel('Time ( ms)')
plt.show()
```
